```python
# load the dataset

from datasets import load_from_disk

items  = load_from_disk("/home/pateisreal/Desktop/dataset/item_dataset_final_top2000")
reviews = load_from_disk("/home/pateisreal/Desktop/dataset/review_dataset_final_top2000")

# convert to pandas
items_df = items.to_pandas()
reviews_df = reviews.to_pandas()

```

```python
reviews_df = reviews_df.rename(columns={"parent_asin": "item_id"})
ratings_df = reviews_df[["user_id", "item_id", "rating"]].copy()
```

```python
items_df = items_df.rename(columns={"parent_asin": "item_id"})
items_df = items_df[[
    "item_id",
    "title",
    "features",
    "description",
    "image_url"
]]
```

```python
products = set(items_df["item_id"].unique())
ratings_df = ratings_df[ratings_df["item_id"].isin(products)]
data = ratings_df.merge(items_df, on="item_id", how="inner")
data.head(5)
```

```python
data.to_csv("/home/pateisreal/projects/ML_project/ML-project/data/processed/products.csv", index=False)
```
